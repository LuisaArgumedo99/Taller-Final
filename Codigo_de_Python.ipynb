```python
# Simulación de un sistema de gestión de usuarios y productos

# Definición de usuarios iniciales
usuarios = {
    "Ren Tao": {"password": "1234_4321", "rol": "admin"},
    "Ioh Asakura": {"password": "abc_cba", "rol": "vendedor"},
    "Chocolove McDonnell": {"password": "11110000", "rol": "vendedor"}
}

# Definición de productos iniciales
productos = {
    "1": {"nombre": "Ladrillo 10x20x30", "precio_compra": 1200.00, "precio_venta": 1500.00, "cantidad": 1000, "unidad": "un"},
    "2": {"nombre": "Cemento Gris x50kg", "precio_compra": 28000.00, "precio_venta": 30000.00, "cantidad": 200, "unidad": "saco"},
    "3": {"nombre": "Varilla Corrugada #5 x6m", "precio_compra": 33000.00, "precio_venta": 35000.00, "cantidad": 150, "unidad": "un"},
    "4": {"nombre": "Impermeabilizante 22kg", "precio_compra": 350000.00, "precio_venta": 360000.00, "cantidad": 50, "unidad": "un"},
    "5": {"nombre": "Malla Electro Soldada 6x2.2m", "precio_compra": 42000.00, "precio_venta": 45000.00, "cantidad": 75, "unidad": "un"}
}

# Lista para registrar las ventas realizadas
ventas = []

def main():
    print("Bienvenido a la TIENDA SHAMAN TOOLS")
    while True:
        print("\nMenú Principal")
        print("1. Iniciar Sesión")
        print("2. Salir")

        opcion = int(input("Seleccione una opción: "))
        if opcion == 2:
            print("Cerrando el sistema...")
            break
        elif opcion == 1:
            login()
        else:
            print("Opción no válida. Por favor, seleccione una opción correcta.")

def login():
    while True:
        nombre = input("Ingrese nombre de usuario: ")
        contraseña = input("Ingrese contraseña: ")

        if verificar_credenciales(nombre, contraseña):
            rol = obtener_rol(nombre)
            print("Sesión iniciada. Menú según el Rol:")
            if rol == "admin":
                admin_menu()
            else:
                vendedor_menu()
            break
        else:
            print("Error: Credenciales incorrectas. Intente de nuevo.")

def verificar_credenciales(nombre, contraseña):
    return nombre in usuarios and usuarios[nombre]["password"] == contraseña

def obtener_rol(nombre):
    return usuarios[nombre]["rol"]

def admin_menu():
    while True:
        print("\nMenú Admin")
        print("1. Registrar usuario")
        print("2. Agregar/Actualizar producto")
        print("3. Consultar inventario")
        print("4. Comprar productos")
        print("5. Ver resumen de ventas")
        print("6. Salir")

        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            registrar_usuario()
        elif opcion == 2:
            agregar_actualizar_producto()
        elif opcion == 3:
            consultar_inventario()
        elif opcion == 4:
            comprar_productos()
        elif opcion == 5:
            ver_resumen_ventas()
        elif opcion == 6:
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción correcta.")

def vendedor_menu():
    while True:
        print("\nMenú Vendedor")
        print("1. Consultar productos")
        print("2. Vender productos")
        print("3. Salir")

        opcion = int(input("Seleccione una opción: "))
        if opcion == 1:
            consultar_inventario()
        elif opcion == 2:
            vender_productos()
        elif opcion == 3:
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción correcta.")

def registrar_usuario():
    nuevo_nombre = input("Ingrese nombre del nuevo usuario: ")
    nueva_contraseña = input("Ingrese contraseña del nuevo usuario: ")
    nuevo_rol = input("Ingrese rol del nuevo usuario (admin/vendedor): ").lower()

    if nuevo_rol not in ["admin", "vendedor"]:
        print("Error: Rol inválido. Debe ser 'admin' o 'vendedor'.")
        return

    if nuevo_nombre in usuarios:
        print("Error: El usuario ya existe.")
    else:
        usuarios[nuevo_nombre] = {"password": nueva_contraseña, "rol": nuevo_rol}
        print("Usuario creado exitosamente.")

def agregar_actualizar_producto():
    codigo = input("Ingrese código del producto: ")
    nombre = input("Ingrese nombre del producto: ")
    precio_compra = float(input("Ingrese precio de compra del producto: "))
    precio_venta = float(input("Ingrese precio de venta del producto: "))
    cantidad = int(input("Ingrese cantidad del producto: "))

    if codigo in productos:
        productos[codigo]["cantidad"] += cantidad
        print("Cantidad del producto actualizada.")
    else:
        productos[codigo] = {"nombre": nombre, "precio_compra": precio_compra, "precio_venta": precio_venta, "cantidad": cantidad, "unidad": "un"}
        print("Producto agregado al inventario.")

def consultar_inventario():
    codigo = input("Ingrese código del producto (o presione '0' para listar todos): ")

    if codigo == "0":
        print("\nInventario completo:")
        for codigo, detalles in productos.items():
            print(f"{codigo}: {detalles['nombre']} | Precio Compra: {detalles['precio_compra']} | Precio Venta: {detalles['precio_venta']} | Cantidad: {detalles['cantidad']}")
    elif codigo in productos:
        detalles = productos[codigo]
        print(f"Detalles del producto {codigo}: {detalles}")
    else:
        print("Producto no encontrado.")

def comprar_productos():
    codigo = input("Ingrese código del producto: ")
    cantidad = int(input("Ingrese cantidad a comprar: "))

    if codigo in productos:
        productos[codigo]["cantidad"] += cantidad
        print("Cantidad de producto actualizada.")
    else:
        nombre = input("Ingrese nombre del nuevo producto: ")
        precio_compra = float(input("Ingrese precio de compra del nuevo producto: "))
        precio_venta = float(input("Ingrese precio de venta del nuevo producto: "))
        productos[codigo] = {"nombre": nombre, "precio_compra": precio_compra, "precio_venta": precio_venta, "cantidad": cantidad, "unidad": "un"}
        print("Producto agregado al inventario.")

def ver_resumen_ventas():
    print("\nResumen de ventas:")
    for venta in ventas:
        print(f"Código: {venta['codigo']} | Cantidad vendida: {venta['cantidad']} | Total generado: {venta['total']}")

def vender_productos():
    codigo = input("Ingrese código del producto: ")
    cantidad = int(input("Ingrese cantidad a vender: "))

    if codigo in productos and productos[codigo]["cantidad"] >= cantidad:
        total = cantidad * productos[codigo]["precio_venta"]
        productos[codigo]["cantidad"] -= cantidad
        ventas.append({"codigo": codigo, "cantidad": cantidad, "total": total})
        print("Venta realizada con éxito.")
    else:
        print("Stock insuficiente para realizar la venta.")

# Inicia el sistema de inventario
main()
```
